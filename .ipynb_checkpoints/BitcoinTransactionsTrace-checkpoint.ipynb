```python
import requests
import pprint
import json
import codecs
import time
from websocket import create_connection
import csv as csv 
```

```python
pp = pprint.PrettyPrinter(indent=4)

def get_lastest_block():
#     url = 'https://bitnodes.21.co/api/v1/snapshots/' + TIMESTAMP + '/'
    url = 'https://blockchain.info/latestblock'
    response = requests.get(url)
    data = response.json()
#     IP = data[][0]
    return data
```

```python
lastest_block = get_lastest_block()
```

```python
for txIdx in lastest_block["txIndexes"]:
    url = 'https://blockchain.info/tx-index/' + str(txIdx) + '?format=json'
    response = requests.get(url)
    data = response.json()
    IP = data["relayed_by"]
    # from IP construct lat and lng
    if IP == '0.0.0.0' or IP == '127.0.0.1':
        continue
    else:
#         print IP
        response = requests.get('http://ip-api.com/json/' + IP)
        loc_data = response.json()
        print loc_data["lat"], loc_data["lon"]
        time.sleep(0.41)
```

# Use websocket 
## the IP geolocation service is prone to reject service

```python
# create saved file
tx_local_data_file = open("tx_local_data.csv", "wb")
tx_local_data_w = csv.writer(tx_local_data_file)
tx_local_data_w.writerow(['lat', 'lon', 'city', 'country'])
# create WebSocket
ws = create_connection("wss://ws.blockchain.info/inv")
ws.send('{"op":"unconfirmed_sub"}')
for i in xrange(1000000):
    result =  ws.recv()
    tx = json.loads(result)
    IP = tx['x']['relayed_by']
    response = requests.get('http://ip-api.com/json/' + IP)
    loc_data = response.json()
    try:
        # print loc_data["lat"], loc_data["lon"], loc_data["city"], loc_data["country"]
        tx_local_data_w.writerow([loc_data["lat"], loc_data["lon"], loc_data["city"], loc_data["country"]])
    except:
        continue
    time.sleep(0.41)
    tx_local_data_file.flush()
ws.close()
tx_local_data_file.close()
```

# alternative IP geolocation APIs - 2500queries/day

```python
# create saved file
tx_local_data_file = open("tx_local_data.csv", "wb")
tx_local_data_w = csv.writer(tx_local_data_file)
tx_local_data_w.writerow(['lat', 'lon', 'city', 'country'])
# create WebSocket
ws = create_connection("wss://ws.blockchain.info/inv")
ws.send('{"op":"unconfirmed_sub"}')
for i in xrange(1000000):
    result =  ws.recv()
    tx = json.loads(result)
    IP = tx['x']['relayed_by']
    response = requests.get('http://freegeoip.net/json/' + IP)
    loc_data = response.json()
    if (loc_data["latitude"] == 0 and loc_data["longitude"] == 0):
        continue
    try:
        # print loc_data["lat"], loc_data["lon"], loc_data["city"], loc_data["country"]
        tx_local_data_w.writerow([loc_data["latitude"], loc_data["longitude"], loc_data["city"], loc_data["country_name"]])
    except:
        continue
    time.sleep(0.3)
    tx_local_data_file.flush()
ws.close()
tx_local_data_file.close()
```
